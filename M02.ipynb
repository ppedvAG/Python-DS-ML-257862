## reshape ##

Form eines Arrays verändern (1D -> 2D, 2D -> 1D)

```python
h = np.random.randint(10, size=(10,5))
```

```python
h
```

```python
h.reshape(5,10)
```

```python
h.reshape(50)
```

```python
h.reshape(h.size)
```

```python
h.reshape(-1)
```
