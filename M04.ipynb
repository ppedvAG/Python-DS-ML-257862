#### N.A.

Keine automatische Lösung -> manuell

Via. String Replace


```python
data == "N.A."
```

```python
(data == "N.A.").value_counts()
```

Spaltenweise machen -> Spalten durchgehen, prüfen wo N.A. ist -> mit loc austauschen

```python
for col in data.columns:
    x = data[col] == "N.A."
    data.loc[x, col] = 0 # inplace = True brauchen wir nicht, weil wir den Wert direkt mit "=" überschreiben
```

```python
data
```

```python
(data == "N.A.").value_counts()
```

##### Funktion fixNA

```python
def fixNA(colName:str, newValue):
    x = data[colName] == "N.A."
    data.loc[x, colName] = newValue
```

```python
# fixNA("FertRate", 1)
```

```python
for col in data.columns:
    fixNA(col,0)
```

#### Prozentzeichen
Muss mithilfe von String Replace gemacht werden

```python
data["YearlyChange"] = data["YearlyChange"].replace(" %", "", regex=True)
```

```python
data
```

##### FUnktion fixPct

```python
def fixPct(colName):
    x = data[colName].replace(" %", "", regex=True)
    data[colName] = x
```

```python
fixPct("UrbanPopPct")
```

```python
fixPct("WorldSharePct")
```

```python
data
```

```python
def fixType(colName, newType):
    data[colName] = data[colName].astype(newType)
```
