### read_csv

Daten werden aus einer CSV-Quelle eingelesen

Wird per Parameter konfiguriert

Beispiele
- delimiter
- thousands
- decimal
- pare_dates
- index_col
- ...

https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html

```python
data = pd.read_csv("Data/PopulationData.csv", delimiter=";", thousands=",", decimal=".")
```

```python
data
```

```python
data.info() # informationen zu spalten
```

### Probleme mit dem Datenset

### Index setzen
set_index("Name") ODER index_col bei read_csv(...)

WICHTIG: Jeder Befehl erzeugt immer eine Kopie des Datensets (mit Änderung)

Zwei Optionen:
- Bestehende Variablen überschreiben
- inplace

```python
data.set_index("Country (or dependency)", inplace=True)
```

```python
data
```

### Spalten anpassen

Spalten entfernen: drop(...)

Spalten umbenennen: rename(...)

rename benötigt ein Dictionary mit den colums=Parameter

```python
data.index.name="Country"
```

```python
data
```

```python
data.drop(columns=["#"], inplace=True)
```

```python
data
```

```python
data.rename(columns={
    "Population(2020)": "Population",
    "Density(P/Km²)": "Density",
    "Land Area(Km²)": "LandArea",
    "Migrants(net)": "Migrants",
    "Fert.Rate": "FertRate",
    "Med.Age": "MedAge",
    "UrbanPop %": "UrbanPopPct",
    "WorldShare": "WorldSharePct"
}, inplace=True)
```

```python
data
```

#### Sortieren

sort_index()

sort_values(["S1", "S2"])

Alle Sortierungen können auch mit ascending=True/False auf auf- oder absteigend gesetzt werden.

```python
data.sort_index() #sortiert index aufsteigend (bei Text nach Alphabet)
```

```python
data.sort_index(ascending=False)
```

```python
data.sort_values("Population", ascending=False)
```

```python
data.sort_values(["LandArea", "Population"], ascending=False)
```

```python
data.sort_values(["Density", "LandArea"], ascending=[False, True])
```

#### Auf Daten zugreifen

Drei Möglichkeiten
- Index: Zugriff auf Zeilen/Spalten Kombination
- loc: Mehrere Spalten
- iloc: Wie loc, aber nur mit Zahlen

```python
data["Population"] # einzelne Spalte anzeigen
```

```python
data["China":"Russia"] # Mehrere Zeilen anzeigen
```

```python
data[0:9] # mehrere Zeilen mit Zahlen
```

```python
data["Population"]["China"] # einzelnen Wert anzeigen
```

```python
data["Population"]["China":"Russia"] # mehrere Werte einer Spalte
```

```python
data.loc["China":"Russia", "Population":"Density"] # mehrere Spalten anzeigen
```

```python
data.loc["China":"Russia", ("Population", "Density")] # Einzelne Spalten selektieren
```

```python
data.loc[:,("Population", "Density")]
```

```python
data.loc[("China", "Russia"), :] # Einzelne Zeilen selektieren
```
