## Serie
Ein Numpy Array mit extra Eigenschaften

- Name
- Benannter Index
- und mehr

Auf das unterliegende numpy Array kann immer zugegriffen werden

```python
e = pd.Series([8.3, 2.3, 1.7, 9.9])
```

```python
e
```

SDiese Zahlen alleine sagen nicht viel aus.


### Name

```python
e.name = "Einwohnerzahlen (Millionen)"
```

```python
e
```

### Index

```python
e.index = ["CH", "DE", "AT", "IT"]
```

```python
e
```

Auf Werte Ã¼ber Index zugreifen

```python
e["CH"]
```

```python
e["CH":"DE"] # Obergrenze inkludiert
```

```python
e[0:2] # Obergrenze exkludiert
```

Zugriff auf das unterliegende Numpy Array

```python
e.values
```

### Vektorisierung von Serien

```python
e>5
```

```python
e[e>5]
```

```python
e.mean()
```

```python
e > e.mean()
```

```python
e[e>e.mean()]
```
